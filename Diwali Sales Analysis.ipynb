```python
# Import Libraries
import pandas as pd
import numpy as np
# For Visulizations
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import Data Sets From Computer
sales = pd.read_csv("Diwali Sales Data.csv",encoding = "unicode_escape")
sales.head(5)
```

```python
# Shape of data
sales.shape
```

```python
# Droping empty columns
sales.drop(columns = ["Status","unnamed1"],inplace = True)
```

```python
# Checking Null Values
sales.isnull().sum()
```

```python
# Null Values less than 5% i.e we can drop them
sales.dropna(inplace = True)
```

```python
# Checking for null values again after droping them
sales.isnull().sum()
```

```python
sales.head()
```

```python
# Data Info
sales.info()
```

```python
# Changing data types
sales["Amount"] = sales["Amount"].astype('int')
sales["Amount"].dtypes
```

```python
# Renaming Column
sales.rename(columns = {"Orders":"Orders Quantity"},inplace = True)
```

```python
sales.head()
```

```python
# Columns Names
sales.columns
```

```python
# Checking the description of the data
sales.describe()
```

# EDA...!

```python
# Checking Corr
sales.corr()
```

```python
plt.figure(figsize=(14,10))
sns.heatmap(sales.corr(), annot = True, cmap="Greens",annot_kws = {"size": 15})
plt.xticks(rotation = 42,size = 15)
plt.yticks(rotation = 42,size = 15)
plt.show()
```

# Gender

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x = "Gender",data = sales,hue=sales["Gender"])
# To Add Values Count 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Spend Ammount
sales_gen = sales.groupby("Gender")["Amount"].sum().reset_index().sort_values(by = "Amount", ascending = True)
# Ploting bar plot
plt.figure(figsize=(10,5))
sns.barplot(x = "Gender",y = "Amount",data = sales_gen,hue = "Amount")
plt.show()
```

### From above graphs we can see that most of the buyers are female and even the purchasing power of females are greater than male

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x = "Age Group",data = sales,hue=sales["Gender"])
# To Add Values Count 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Spend Ammount
sales_age = sales.groupby("Age Group")["Amount"].sum().reset_index().sort_values(by = "Amount", ascending = True)
plt.figure(figsize=(10,5))
sns.barplot(x = "Age Group", y = "Amount",data = sales_age)
plt.show()
```

### From above graph we can see that most of the buyers are of age group between 26-35 yrs. female

# State

```python
# Total no. of orders from top 10 states
sales_state = sales.groupby("State")["Orders Quantity"].sum().reset_index().sort_values(by = "Orders Quantity", ascending = False).head(10)

# Plot
sns.set(rc = {"figure.figsize":(15,10)})
sns.barplot(data = sales_state,x = "State",y = "Orders Quantity")
plt.show()
```

```python
# Total amount / Sales from top 10 states
sales_state = sales.groupby("State")["Amount"].sum().reset_index().sort_values(by = "Amount", ascending = False).head(10)
# Plot
sns.set(rc = {"figure.figsize":(15,10)})
sns.barplot(data = sales_state,x = "State",y = "Amount")
plt.show()
```

### From above graphs we can see that most of the orders & total sales/ amount are from UP,MH,Karnataka respectively

# Martial Status

```python
# Purchasing Based on Maritial Status
plt.figure(figsize=(10,5))
ax = sns.countplot(x = 'Marital_Status',data = sales,hue=sales["Gender"])
# To Add Values Count 
for bars in ax.containers:
    ax.bar_label(bars)
```

### From above graph we can see that most of the buyers are married (Women) and they have high purchasing power

# Occupation

```python
plt.figure(figsize=(14,5))
ax = sns.countplot(data = sales, x = "Occupation")
plt.xticks(rotation = 45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
purchaing = sales.groupby("Occupation")["Amount"].sum().reset_index().sort_values(by = "Amount", ascending = False).head(10)
plt.figure(figsize=(14,5))
sns.barplot(data = purchaing, x = "Occupation", y = "Amount")
```

#### From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation Sectores

# Product Category

```python
plt.figure(figsize=(14,7))
ax = sns.countplot(data = sales, x = "Product_Category")
plt.xticks(rotation = 45)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

```python
purchaing = sales.groupby("Product_Category")["Amount"].sum().reset_index().sort_values(by = "Amount", ascending = False).head(10)
plt.figure(figsize=(14,5))
sns.barplot(data = purchaing, x = "Product_Category", y = "Amount")
plt.xticks(rotation = 45)
plt.show()
```

#### From graphs we can see that most of the  sold products are from food clothing and Electronics category

## Conclusion :

#### Married women age group 26 -35 yrs from Up, MH and Karnataka working in IT, Healthcare and aviation are more likely buy products from food clothing and electronics category
